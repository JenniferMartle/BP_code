#### SUBJECTS

```python
df['genres'].fillna(',', inplace=True)
```

```python
df['genres'] = df.apply(lambda r: r['genres'].split(',')[1:], 1)
```

```python
df.head(5)
```

```python
df['subjects'].fillna('', inplace=True)
```

```python
df['subjects'] = df.apply(lambda r: r['subjects'].split(','), 1)
```

```python
dfS=df[['genres', 'subjects']]
```

```python
dfS[dfS['subjects'].str.get(0)==''].head()
```

```python
dfS['genres']=dfS['genres'].mask(dfS['genres'].str.get(0)=='')
dfS['subjects']=dfS['subjects'].mask(dfS['subjects'].str.get(0)=='')
```

```python
dfS=dfS.bfill(axis='columns')
```

```python
dfS.drop('subjects', axis=1, inplace=True)
df.drop(['subjects', 'genres'], axis=1, inplace=True)
dfS.head()
```

```python
df=df.merge(dfS, left_index=True, right_index=True)
df.head()
```

```python
len(df[df['genres'].isna()])
```

```python
df['genres']=df['genres'].fillna('unknown').apply(list)
```

#### AUTHOR

```python
len(df[(df['creator'].isna())&(df['author'].notna())])
```

```python
dfC=df[['creator', 'author']]
```

```python
dfC['creator'].fillna('creator, unknown', inplace=True)
```

```python
dfC['creator'] = dfC.apply(lambda r: r['creator'].split(', ')[:2], 1)
```

```python
dfC.tail()
```

```python
dfC['creator'][0][1]
```

```python
def first_name_select(name):
    if len(name)>1:
        return name[1]
    else:
        return ''
```

```python
dfC['first_name']=dfC['creator'].apply(first_name_select)
dfC['last_name']=dfC.apply(lambda r: r['creator'][0], 1)
```

```python
dfC.tail()
```

```python
dfC['creator']=dfC['first_name']+' '+dfC['last_name']
dfC.drop(['first_name', 'last_name'], axis=1, inplace=True)
```

```python
dfC['creator']=dfC['creator'].mask(dfC['creator']=='unknown creator')
```

```python
dfC=dfC.bfill(axis='columns')
```

```python
dfC['creator'].fillna('unknown', inplace=True)
```

```python
dfC.drop('author', axis=1, inplace=True)
df.drop(['creator', 'author'], axis=1, inplace=True)
```

```python
df=df.merge(dfC, left_index=True, right_index=True)
df.head()
```
